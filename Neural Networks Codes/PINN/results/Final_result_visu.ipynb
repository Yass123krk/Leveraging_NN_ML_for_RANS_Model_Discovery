This Notebook file presents the results on unseen data obtained using the models that have been created.

There are two main models:

1. Unified Model: This model is designed to predict both Channel and Couette flow.
   
2. Splitted Model: This model is designed to predict only one type of flow. Therefore, two models have been built: one for predicting the Channel flow and another one for predicting the Couette flow.

It's important to note that each of these models exists in two different types:

- RANS-based PINN
- YSINDy-based PINN

## RANS based PINN Couette

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the CSV files for the following columns: u'u',v'v',w'w',u'w',u'v',v'w'
ground_truth_df = pd.read_csv(r"C:\Users\Sohai\OneDrive - Cranfield University\GroupProject\Neural-Network\DNS_data\Couette_data\mirrored_LM_Couette_R0500_100PI_prof.csv")
predicted_df = pd.read_csv(r"C:\Users\Sohai\OneDrive - Cranfield University\GroupProject\Neural-Network\Result_comparison\Result_RANS_mirrored_LM_Couette_R0500_100PI_prof.csv")

# for the x-axis
y = ground_truth_df['y^+']

# Create a figure and set of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 4))
axs = axs.flatten()

columns = ["u'u'", "v'v'", "w'w'", "u'v'"]
 
# Plot variances and covariance result
for i, col in enumerate(columns):
    # Ground truth plot
    axs[i].plot(y, ground_truth_df[col], label='Ground Truth', color='blue')
    # Predicted plot
    axs[i].plot(y, predicted_df[col], label='Predicted', color='red', linestyle='--')
    axs[i].set_title(col)
    axs[i].legend()

plt.suptitle("mirrored_LM_Couette_0500_100PI")
plt.tight_layout()

plt.show()
```
